```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
def create_neural_network(input_dim, hidden_units, learning_rate=0.01):
    """
    Create a neural network model with a specified number of hidden units.

    Args:
        input_dim (int): Number of input features.
        hidden_units (int): Number of units in the hidden layer.
        learning_rate (float): Learning rate for the Adam optimizer.

    Returns:
        model (Sequential): Compiled neural network model.
    """
    model = Sequential([
        Dense(hidden_units, input_dim=input_dim, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
def train_neural_network(model, X_train, y_train, num_epochs, batch_size):
    """
    Train the neural network model on the training data.

    Args:
        model (Sequential): Compiled neural network model.
        X_train (ndarray): Training input features.
        y_train (ndarray): Training labels.
        num_epochs (int): Number of training epochs.
        batch_size (int): Batch size for training.

    Returns:
        history: Keras History object with training metrics.
    """
    history = model.fit(
        X_train, y_train,
        epochs=num_epochs,
        batch_size=batch_size,
        verbose=1
    )
    return history
```

```python
# Example data and labels
np.random.seed(42)
X_train = np.random.rand(100, 4)
y_train = np.random.randint(0, 2, size=(100,))
```

```python
# Hyperparameters to experiment with
learning_rates = [0.001, 0.01, 0.1]
hidden_units_list = [4, 8, 16]
```

```python
# Perform hyperparameter tuning
for lr in learning_rates:
    for hidden_units in hidden_units_list:
        print("=" * 60)
        print(f"Experiment | Learning Rate: {lr}  |  Hidden Units: {hidden_units}")
        print("=" * 60)

        model = create_neural_network(
            input_dim=X_train.shape[1],
            hidden_units=hidden_units,
            learning_rate=lr
        )

        history = train_neural_network(
            model=model,
            X_train=X_train,
            y_train=y_train,
            num_epochs=100,
            batch_size=16
        )

        final_loss = history.history['loss'][-1]
        final_accuracy = history.history['accuracy'][-1]
        print(f"\nResults -> Loss: {final_loss:.4f}  |  Accuracy: {final_accuracy:.4f}\n")
```
